```python
#Basic python libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
#Loading the dataset
df = sns.load_dataset('car_crashes')
```

```python
#Printing the head of data
df.head()
```

```python
#Printing the shape of datset
df.shape
```

```python
df.info()
```

```python
sns.heatmap(df.corr(), annot = True)
```

```python
#Bar plot between cities and total no of accidents
plt.figure(figsize = (15,5))
sns.barplot(x = 'abbrev', y = 'total', data = df)
```

```python
df1 = df[['abbrev','total']].sort_values(by='total').tail(15)
df1.shape
```

```python
#Bar plot of 10 cities with maximum accidents
plt.figure(figsize = (7,5))
sns.barplot(x='abbrev', y='total', data = df1)
plt.xlabel('State')
plt.ylabel('Total')
plt.title('Total number of accidents occured in United States')
```

```python
df
```

```python
df2 = df[['abbrev','ins_losses','total']].sort_values(by='total').tail(15)
df2
```

```python
#Plotting the graph 15 top most cities with maximum no of accidents
plt.figure(figsize = (10,5))
sns.barplot(y = 'ins_losses', x = 'abbrev', data = df2, palette = 'viridis')
plt.xlabel('States')
plt.ylabel('Insurance loss')
plt.title('Number of insurance loss in the states as the total number of accidents are increasing')
```

```python
df3 = df[['abbrev','ins_premium','total']].sort_values(by='total').tail(5)
df4 = df[['abbrev','ins_premium','total']].sort_values(by='total').head(5)
```

```python
df3
```

```python
df4
```

```python
df5 = pd.concat([df4, df3])
```

```python
df5
```

```python
#Plotting graph of 5 highest and 5 lowest total no of accidents occured in the cities
plt.figure(figsize = (7,6))
sns.barplot(x = 'abbrev', y = 'ins_premium', data = df5, palette = 'coolwarm')
plt.xlabel('State')
plt.ylabel('Premium insurance')
plt.title('Insurance where total number of incidents are maximum and minimumin in states')
```

```python
#Scatter plot of total no of accidents with premium insurance and losses
plt.figure(figsize = (12,4))
plt.tight_layout()
plt.subplot(1,2,1)
sns.scatterplot(x= 'ins_premium', y = 'total', data = df, color = 'orange')
plt.xlabel('Premium insurance')
plt.ylabel('Total')
plt.title('Plot betewen premium insurance and total no. of accidents')
plt.subplot(1,2,2)
sns.scatterplot(x= 'ins_losses', y = 'total', data = df, color = 'red')
plt.xlabel('Insurance losses')
plt.ylabel('Total')
plt.title('Plot between insurance losses and total no. of accidents')
```

```python
sns.scatterplot(x ='total', y = 'not_distracted', data = df)
```

```python
#Graph of total no of accidents happened due to any reason
df6 = df[['total','not_distracted','speeding','alcohol','no_previous']].sort_values(by='total').tail(18)
plt.figure(figsize = (17,10))
sns.pointplot(x = 'total', y = 'not_distracted', data = df6, color = 'red', alpha = 0.3)
sns.pointplot(x = 'total', y = 'speeding', data = df6, color = 'blue')
sns.pointplot(x = 'total', y = 'alcohol', data = df6, color = 'green')
sns.pointplot(x = 'total', y = 'no_previous', data = df6, color = 'orange')
plt.title('Number of accidents happened due to any reason')
plt.grid()
plt.text(2,0,'speeding',color='blue',fontsize = 15,style = 'italic')
plt.text(4,0,'alcohol',color='green',fontsize = 15,style = 'italic')
plt.text(6,0,'Not Distracted',color='red',fontsize = 15,style = 'italic')
plt.text(9,0,'No Previous Accident',color='orange',fontsize = 15,style = 'italic')
plt.xlabel('Total')
plt.ylabel('Count')
```
